```python
import sys
import commands as cmd
import MeCab 

m = MeCab.Tagger()
o = MeCab.Tagger("-Owakati")
```

```python
def checkhinsi(line):
    print "-**************************************"
    print "line: " + line
    
    ### pline: mecab による解析結果が入った str 型
    pline = (m.parse(line))
    ### pline2: 単語の解析結果を要素にもつ list 型に
    pline2 = pline.split("\n")
    ### wakati: mecab による分かち書き形式に変換したline. str 型
    wakati = o.parse(line)
    ### words: line を単語要素とした list 型
    words = wakati.split(" ")
    
    print "num of words: ",
    print len(words) - 1
    
    ### phrase 自立語と付属語をそれぞれ 1, 0 として line を表現
    phrase = []
    
    for i in range(0,len(words)-1):
        if "名詞" in pline2[i]:
            print "名詞: " + words[i];
            phrase.append("1")
        elif "動詞" in pline2[i]:
            print "動詞: " + words[i];
            phrase.append("1")
        elif "形容詞" in pline2[i]:
            print "形容詞: " + words[i];
            phrase.append("1")
        elif "副詞" in pline2[i]:
            print "副詞: " + words[i];
            phrase.append("1")
        elif "助詞" in pline2[i]:
            print "助詞: " + words[i];
            phrase.append("0")
        elif "接続詞" in pline2[i]:
            print "接続詞: " + words[i];
            phrase.append("1")
        elif "助動詞" in pline2[i]:
            print "助動詞: " + words[i];
            phrase.append("0")
        elif "連体詞" in pline2[i]:
            print "連体詞: " + words[i];
            phrase.append("1")
        elif "感動詞" in pline2[i]:
            print "感動詞: " + words[i];
            phrase.append("1")
        elif "記号" in pline2[i]:
            print "記号: " + words[i];
            phrase.append("1")
        else:
            print "未知語: " + words[i];
            phrase.append("2")
        #elif "" in pline:
        #    print ": " + line
        
    print ;
    print phrase;
    
    split2phrase(words, phrase)
    
    #temp = cmd.getstatusoutput("cat " + pline + "| awk '{print $2}' | sed -e 's/,.*//g'" )
```
