```python
import sys
import commands as cmd
import MeCab 

m = MeCab.Tagger()
o = MeCab.Tagger("-Owakati")
```

```python
def checkhinsi(line):
    print "-**************************************"
    print "line: " + line
    ### pline mecab による解析結果が入った str 型
    pline = (m.parse(line))
    ### pline 単語の解析結果を要素にもつ list 型に
    pline2 = pline.split("\n")
    ### wakati mecab による分かち書き形式に変換したline. str 型
    wakati = o.parse(line)
    ### words line を単語要素とした list 型
    words = wakati.split(" ")
    print "num of words: ",
    print len(words)
    for i in range(0,len(words)):
        if "名詞" in pline2[i]:
            print "名詞: " + words[i];
        elif "動詞" in pline2[i]:
            print "動詞: " + words[i];
        elif "形容詞" in pline2[i]:
            print "形容詞: " + words[i];
        elif "副詞" in pline2[i]:
            print "副詞: " + words[i];
        elif "助詞" in pline2[i]:
            print "助詞: " + words[i];
        elif "接続詞" in pline2[i]:
            print "接続詞: " + words[i];
        elif "助動詞" in pline2[i]:
            print "助動詞: " + words[i];
        elif "連体詞" in pline2[i]:
            print "連体詞: " + words[i];
        elif "感動詞" in pline2[i]:
            print "感動詞: " + words[i];
        elif "記号" in pline2[i]:
            print "記号: " + words[i];
        #elif "" in pline:
        #    print ": " + line
    
    temp = cmd.getstatusoutput("cat " + pline + "| awk '{print $2}' | sed -e 's/,.*//g'" )
    #print i;
    #print temp
```
